```python
import numpy as np
```

```python
mag = np.random.normal(size=(10, 5))
pha = np.random.uniform(0, 2*np.pi, size=(10, 5))
```

```python
cmplx = mag * np.exp(-pha)
```

```python
out = np.fft.irfftn(cmplx)
```
